```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
db = pd.read_csv("trees.csv")
db.head()
```

```python
plt.scatter(x=db.Girth,y=db.Volume, color = "red", facecolors="none", marker="*")
plt.xlabel("Circuferencia")
plt.ylabel("Volume")
plt.title("Arvores")
```

```python
plt.plot(db.Girth, db.Volume)
plt.title("Arvores")
plt.xlabel("Circuferencia")
plt.ylabel("Volume")
```

```python
sns.regplot(x=db.Girth, y=db.Volume, data = db, x_jitter=0.3, fit_reg=True)
```

```python
#GRAFICOS DE DISPERSAO COM LEGENDAS

db2 = pd.read_csv("co2.csv")
db2.head()
```

```python
x=db2.conc
y=db2.uptake

```

```python
unicos = list(set(db2.Treatment))
unicos
```

```python


for unico in unicos:
    idx = db2.Treatment == unico
    plt.scatter(x[idx], y[idx], label=unico)
plt.legend(loc = "lower right")

```
