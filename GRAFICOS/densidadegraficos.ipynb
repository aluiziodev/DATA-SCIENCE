```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
db = pd.read_csv("trees.csv")
db.shape
```

```python
#HISTOGRAMA

plt.hist(db.iloc[:,1], bins=6)
```

```python
#HISTOGRAMA SEABORN

sns.histplot(db.iloc[:,1], kde=False, bins=6, color="red").set_title("Arvores")
```

```python
#LINHA DE DENSIDADE

sns.kdeplot(db.iloc[:,1], color="red").set_title("Arvores")
```

```python
#LINHA SOBREPOSTA NO HISTOGRAMA
sns.histplot(db.iloc[:,1], kde=True, bins=6, color="red").set_title("Arvores")
```
