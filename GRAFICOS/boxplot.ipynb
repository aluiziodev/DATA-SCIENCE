```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
db = pd.read_csv("trees.csv")
db.head()
db.shape
```

```python
plt.boxplot(db.Volume, vert=False, showfliers=False, notch= True, patch_artist=True)
plt.xlabel("Volume")
plt.ylabel("Arvores")
```

```python
plt.boxplot(db, vert=False)
plt.xlabel("Dados")
plt.title("Arvore")
```

```python
plt.boxplot(db.Volume, vert=False)
plt.boxplot(db.Girth, vert=False)
plt.boxplot(db.Height, vert=False)
plt.xlabel("Dados")
plt.ylabel("Arvores")
```

```python
#BOXPLOT COM SEABORN

sns.boxplot(data= db.Volume, orient="v").set_title("Arvores")
```
