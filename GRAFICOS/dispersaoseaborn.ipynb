```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
db = pd.read_csv("co2.csv")
db.head()
```

```python
srn.scatterplot(x=db.conc, y=db.uptake, hue= db.Type)
```

```python
q = db.loc[db["Type"]=="Quebec"]
m = db.loc[db["Type"]=="Mississippi"]
```

```python
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(x=q.conc, y=q.uptake).set_title("quebec")
plt.subplot(1,2,2)
srn.scatterplot(x=m.conc, y=m.uptake).set_title("mississippi")
plt.tight_layout()
```

```python
ch = db.loc[db["Treatment"]=="chilled"]
nc = db.loc[db["Treatment"]=="nonchilled"]
```

```python
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(x=ch.conc, y=ch.uptake).set_title("chilled")
plt.subplot(1,2,2)
srn.scatterplot(x=nc.conc, y=nc.uptake).set_title("nonchilled")
plt.tight_layout()
```
