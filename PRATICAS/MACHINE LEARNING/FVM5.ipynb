```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from keras.models import Sequential 
from keras.layers import Dense, Dropout
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split
import numpy as np
from keras.datasets import mnist
import pandas as pd 
```

```python
db = pd.read_csv("soybean.csv")
db.shape
```

```python
db.head(10)
```

```python
x = db.iloc[:,0:35].values
y = db.iloc[:,35].values
y
```

```python
for i in range(35):
    lencoder = LabelEncoder()
    if type(x[0,i])==str:
        x[:,i] = lencoder.fit_transform(x[:,i])
x
```

```python
lencoder = LabelEncoder()
y = lencoder.fit_transform(y)
y = to_categorical(y, 19)
y
```

```python
x_trein, x_test, y_trein, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
print(len(x_trein), len(x_test), len(y_trein), len(y_test))
```

```python
sc = StandardScaler()
x_trein = sc.fit_transform(x_trein)
x_test = sc.transform(x_test)
len(x_trein[0,:])
```

```python
modelo = Sequential()
modelo.add(Dense(units=50, activation="relu", input_dim=35))
modelo.add(Dropout(0.2))
modelo.add(Dense(units=50, activation="relu"))
modelo.add(Dropout(0.2))
modelo.add(Dense(units=50, activation="relu"))
modelo.add(Dropout(0.2))
modelo.add(Dense(units=19, activation="softmax"))
modelo.summary()
```

```python
modelo.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
modelo.fit(x_trein, y_trein, epochs=1000, validation_data=(x_test, y_test))
```

```python
y_pred = modelo.predict(x_test)
y_pred
```

```python
y_test_matrix = [np.argmax(t) for t in y_test]
y_pred_matrix = [np.argmax(t) for t in y_pred]
confusao = confusion_matrix(y_test_matrix, y_pred_matrix)
confusao
```
