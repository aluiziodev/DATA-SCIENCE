```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot
```

```python
db = pd.read_csv('slr12.csv', sep=';')
db.shape
```

```python
db.head()
```

```python
x = db.iloc[:, 0].values
y = db.iloc[:, 1].values


```

```python
correlaçao = np.corrcoef(x, y)
correlaçao
```

```python
x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(x, y)
```

```python
plt.scatter(x, y, color='blue')
plt.plot(x, model.predict(x), color='red')
plt.xlabel('TAXA ANUAL')
plt.ylabel('INVESTIMENTO INICIAL')
plt.title('REGRESSÃO LINEAR SIMPLES - INVESTIMENTO INICIAL X TAXA ANUAL')
plt.show()

```
