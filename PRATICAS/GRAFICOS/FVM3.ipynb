```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
db = pd.read_csv("dados.csv")
db.head()
```

```python
db.shape
```

```python
db.isnull().sum()
```

```python
sns.boxplot(data=db.PIB, orient="v").set_title("PIB DOS MUNICIPIOS")
```

```python
db.loc[db["PIB"]>200000]
```

```python
sns.boxplot(data=db.VALOREMPENHO, orient="v").set_title("VALOREMPENHO DOS MUNICIPIOS")
```

```python
db.loc[db["VALOREMPENHO"]>1.4*1e6]
```

```python
sns.histplot(db.PIB, kde=True, color="blue", stat="density", bins=10).set_title("PIB")
```

```python
sns.histplot(db.VALOREMPENHO, kde=True, color="blue", stat="density", bins=10).set_title("VALOREMPENHO")
```

```python
sns.scatterplot(x=db.PIB, y=db.VALOREMPENHO)
```

```python
agrp = db.sort_values("PIB").head(10)
agrp = agrp.iloc[:,1:3]
agrp.plot.bar(x="MUNICIPIO", y="PIB", color="red")
```

```python
agrp = db.sort_values("VALOREMPENHO").head(10)
agrp = agrp.iloc[:,[1,3]]
agrp.plot.bar(x="MUNICIPIO", y="VALOREMPENHO", color="gray")
```
