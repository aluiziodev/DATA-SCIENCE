```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential 
from keras.layers import Dense
from keras.utils import to_categorical
import numpy as np
```

```python
db = datasets.load_iris()
prev = db.data
classe = db.target
prev
```

```python
classe_dummy = to_categorical(classe)
classe_dummy
```

```python
x_trein, x_test, y_trein, y_test = train_test_split(prev,
                                                    classe_dummy,
                                                    test_size=0.3,
                                                    random_state=0)
```

```python
modelo = Sequential()

modelo.add(Dense(units=5, input_dim=4)) #primeira camada oculta
modelo.add(Dense(units=4)) #segunda camada oculta
modelo.add(Dense(units=3, activation="softmax")) #terceira camada oculta
```

```python
modelo.summary()
```

```python
modelo.compile(optimizer="adam", loss="categorical_crossentropy",
               metrics=["accuracy"])

modelo.fit(x_trein, y_trein, epochs=1000, validation_data=(x_test, y_test))
```

```python
prev = modelo.predict(x_test)
prev = (prev > 0.5)
prev
```

```python
y_test_matrix = [np.argmax(t) for t in y_test]
y_prev_matrix = [np.argmax(t) for t in prev]

confusao = confusion_matrix(y_test_matrix, y_prev_matrix)
confusao
```
