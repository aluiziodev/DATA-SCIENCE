```python
#agrupamento com k-means
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

```

```python
iris = datasets.load_iris()
#visualizando os dados
unicos, qtd = np.unique(iris.target, return_counts=True)
print (unicos)
print (qtd)
```

```python
cluster = KMeans(n_clusters=3)
cluster.fit(iris.data)
```

```python
centroides = cluster.cluster_centers_
centroides
```

```python
agrupamento = cluster.labels_
agrupamento
```

```python
unicos2, qtd2 = np.unique(agrupamento, return_counts=True)
print (unicos2)
print (qtd2)
```

```python
#geraçao matriz de contigencia para comparar os resultados do k-means com os resultados reais
confusao = confusion_matrix(iris.target,agrupamento)
confusao
```

```python
plt.scatter(iris.data[agrupamento==0,0],iris.data[agrupamento==0,1], 
            color='green', label='setosa')
plt.scatter(iris.data[agrupamento==1,0],iris.data[agrupamento==1,1], 
            color='red', label='versicolor')
plt.scatter(iris.data[agrupamento==2,0],iris.data[agrupamento==2,1],
            color='blue', label='virgica')
plt.legend()
```
