```python
#agrupamento com c-means
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score,confusion_matrix
import skfuzzy
```

```python
iris = datasets.load_iris()
iris.data
```

```python
r = skfuzzy.cmeans(data = iris.data.T, c=3, m=2, error=0.005, maxiter=1000, init=None)
```

```python
previsoes_percent = r[1]
previsoes_percent
```

```python
for x in range(150):
    print(f'flor {x+1}: setosa: {previsoes_percent[0][x]:.4f}, versicolor: {previsoes_percent[1][x]:.4f}, virginica: {previsoes_percent[2][x]:.4f}')
```

```python
previsoes = previsoes_percent.argmax(axis=0)
resultados = confusion_matrix(iris.target,previsoes)
resultados
```
