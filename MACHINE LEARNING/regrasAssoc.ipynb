```python
import pandas as pd
import apyori as apriori
```

```python
dataset = pd.read_csv('transacoes.txt', header=None)
dataset
```

```python
transaçoes = []
for i in range(6):
    transaçoes.append([str(dataset.values[i,j]) for j in range(3)])
transaçoes
```

```python
regras = apriori.apriori(transaçoes, min_support=0.5, min_confidence=0.5, min_length=2 )
```

```python
resultados = list(regras)
resultados
```

```python
resultados2 = [list(x) for x in resultados]
resultados2
```

```python
resultado3 = []
for j in range (len(resultados2)):
    resultado3.append([list(x) for x in resultados2[j][2]])
resultado3
```
