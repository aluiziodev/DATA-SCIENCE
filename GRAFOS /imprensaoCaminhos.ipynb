```python
from igraph import Graph
from igraph import plot
```

```python
grafo = Graph(edges = [(0,2),(0,1),(1,4),(1,5),(2,3),(6,7),(3,7),(4,7),(5,6)],
                       directed = True)
grafo.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo.es['weight'] = [2,1,2,1,2,1,3,1]
plot(grafo,edge_label=grafo.es["weight"] ,bbox=(0,0,300,300))
```

```python
caminho_edge = grafo.get_shortest_paths(0, 7, output="epath")
caminho_edge_index = []
for n in caminho_edge[0]:
    caminho_edge_index.append(n)
caminho_edge_index
```

```python
caminho_vertex = grafo.get_shortest_paths(0, 7, output="vpath")
caminho_vertex_label = []
for n in caminho_vertex[0]:
    caminho_vertex_label.append(grafo.vs[n]["label"])
caminho_vertex_label
```

```python
for v in grafo.vs:
    if v["label"] in caminho_vertex_label:
        v["color"] = "green"
    else:
        v["color"] = "gray"
        
```

```python
for e in grafo.es:
    if e.index in caminho_edge_index:
        e["color"] = "green"
    else:
        e["color"] = "gray"
```

```python
plot(grafo, bbox=(0,0,300,300),edge_label=grafo.es["weight"]  )
```
