```python
from igraph import Graph
from igraph import plot
```

```python
grafo1 = Graph(edges=[(0,1), (1,2), (2,3), (3,0)], directed=True)
grafo1.vs['label'] = range(1, grafo1.vcount()+1)
print(grafo1)
```

```python
plot(grafo1, bbox=(0, 0, 300,300))
```

```python
grafo2 = Graph(edges=[(0,1),(1,2),(2,3),(3,0),(0,3), (3,2),(2,1), (1,0)], directed=True)
grafo2.vs["label"] = range(1, grafo2.vcount()+1)
layout = grafo2.layout_circle()
layout.scale(0.5)
plot(grafo2, layout= layout, bbox=(0, 0, 300,300))
```

```python
#grafo com la√ßo

grafo3 = Graph(edges=[(0,1), (1,2), (2,3), (3,0), (1,1)], directed=True)
grafo3.vs["label"] = range(1, grafo3.vcount()+1)
plot(grafo3, bbox=(0, 0, 300,300))
```

```python
#grafo com vertice isolado

grafo4 = Graph(edges=[(0,1), (1,2), (2,3), (3,0), (1,1)], directed=True)
grafo4.add_vertex(5)
grafo4.vs["label"] = range(1, grafo4.vcount()+1)
plot(grafo4, bbox=(0, 0, 300,300))
```
