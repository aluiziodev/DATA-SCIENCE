```python
from igraph import Graph
from igraph import plot
import matplotlib.pyplot as plt
```

```python
grafo = Graph(directed=False)
grafo.add_vertices(5)
grafo.add_edges([(0,1),(1,2), (2,3),(3,4), (4,0),(0,2), (2,1)])
grafo.add_vertices(2)
grafo.vs["label"] = ["A","B","C","D","E","F","G"]
grafo.vs["name"] = ["A","B","C","D","E","F","G"]
print(grafo.get_adjacency())
```

```python
print(grafo.get_adjacency()[0,])
```

```python
print(grafo.get_adjacency()[0,1])
```

```python
for v in grafo.vs:
    print(v)
```

```python
grafo2 = Graph(directed=True)
grafo2.add_vertices(4)
grafo2.add_edges([(0,1), (2,3),(0,2),(0,3)])
grafo2.vs["label"] = ["fernando", "pedro","jose","antonio"]
grafo2.vs["peso"] = [40,30,30,25]
print(grafo2)
```

```python
for v in grafo2.vs:
    print(v)
```

```python
grafo2.es["tipoAmizade"] = ["amigo", "inimigo", "inimigo", "amigo"]
grafo2.es["weight"] = [1,2,1,3]
print(grafo2)
```
