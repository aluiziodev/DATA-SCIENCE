```python
from igraph import Graph
from igraph import plot
```

```python
grafo2 = Graph(directed=True)
grafo2.add_vertices(4)
grafo2.add_edges([(0,1), (2,3),(0,2),(0,3)])
grafo2.vs["label"] = ["fernando", "pedro","jose","antonio"]
grafo2.vs["peso"] = [40,30,30,25]
grafo2.es["tipoAmizade"] = ["amigo", "inimigo", "inimigo", "amigo"]
grafo2.es["weight"] = [1,2,1,3]
grafo2.vs["color"] = ["blue","red","yellow", "green"]

plot(grafo2, bbox=(0,0,300,300), vertex_color=grafo2.vs["color"])

```

```python
plot(grafo2, bbox=(0,0,300,300), vertex_color=grafo2.vs["color"],edge_width=grafo2.es["weight"])
```

```python
layout = grafo2.layout_auto().scale(0.5)

plot(grafo2, bbox=(0,0,300,300), vertex_color=grafo2.vs["color"],
     edge_width=grafo2.es["weight"], vertex_size= grafo2.vs["peso"],
     edge_curved=0.4, layout=layout)
```

```python
layout = grafo2.layout_auto().scale(0.5)

plot(grafo2, bbox=(0,0,300,300), vertex_color=grafo2.vs["color"],
     edge_width=grafo2.es["weight"], vertex_size= grafo2.vs["peso"],
     edge_curved=0.4, layout=layout, vertex_shape="square")
```
