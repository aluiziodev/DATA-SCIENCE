```python
from igraph import Graph
from igraph import plot
import matplotlib.pyplot as plt
```

```python
#grafo nao direcionado

grafo1 = Graph(edges=[(0,1), (1,2), (2,3), (3,0)], directed=False)
grafo1.vs['label'] = range(1, grafo1.vcount()+1)
print(grafo1)
```

```python
plot(grafo1, bbox=(0,0,300,300))
```

```python
grafo2 = Graph(directed=False)
grafo2.add_vertices(10)
grafo2.add_vertex(16)
grafo2.add_edges([(0,1),(2,2), (2,3), (3,0)])
grafo2.vs["label"] = range(1, grafo2.vcount()+1)
plot(grafo2, bbox=(0,0,300,300))
```

```python
grafo3 = Graph(directed=False)
grafo3.add_vertices(5)
grafo3.add_edges([(0,1),(1,2), (2,3),(3,4), (4,0),(0,2), (2,1)])
grafo3.add_vertices(2)
grafo3.vs["name"] = ["A","B","C","D","E","F","G"]
fig, ax = plt.subplots(figsize=(5,5))
plot(grafo3,target=ax, bbox=(0,0,300,300),vertex_label=grafo3.vs["name"])
plt.show()
```
