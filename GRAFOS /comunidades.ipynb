```python
from igraph import Graph
from igraph import plot
import igraph 
import numpy as np
```

```python
grafo = igraph.load("Grafo.graphml")
print(grafo)
```

```python
#plot(grafo, bbox=(0,0,600,600))
```

```python
comunidades = grafo.clusters()
print(comunidades)
```

```python
comunidades.membership
```

```python
color = comunidades.membership
color = np.array(color)
color *= 20
color = color.tolist()
#plot(grafo, bbox=(0,0,600,600), vertex_color=color)
```

```python
grafo2 = Graph(edges = [(0,2),(0,1),(1,4),(1,5),(2,3),(6,7),(3,7),(4,7),(5,6)],
                       directed = True)
grafo2.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo2.es['weight'] = [2,1,2,1,2,1,3,1]
plot(grafo2, bbox=(0,0,300,300))
```

```python
comunidade2 = grafo2.clusters()
print(comunidade2)
```

```python
color2 = comunidade2.membership
color2 = np.array(color2)
color2 *= 20
color2 = color2.tolist()
plot(grafo2, bbox=(0,0,300,300), vertex_color=color2)
```
