```python
from scipy.stats import t
```

```python
#MEDIA DE SALARIOS = 75.00 p/HORA, AMOSTRA COM 9 FUNCIONARIOS, DESVIO PADRAO = 10.00 p/HORA

#qual a prob do salario ser menor que 80.00 p/HORA?
t.cdf(1.5, 8)
```

```python
#qual a prob do salario ser maior que 80.00 p/HORA?

t.sf(1.5, 8) 
```
