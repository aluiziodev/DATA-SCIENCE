```python
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols 
from statsmodels.stats.multicomp import MultiComparison
```

```python
db = pd.read_csv("anova.csv", sep=";")
db.head()   
```

```python
db.shape
```

```python
db.groupby("Sexo").size()
```

```python
db.groupby("Remedio").size()
```

```python
db.boxplot(by="Remedio", grid=False)
```

```python
modelo1 = ols("Horas ~ Remedio", data=db).fit()
resultado1 = sm.stats.anova_lm(modelo1)
print(resultado1)
```

```python
modelo2 = ols("Horas ~ Sexo * Remedio", data=db).fit()
resultado2 = sm.stats.anova_lm(modelo2)
print(resultado2)
```

```python
mc = MultiComparison(db["Horas"], db["Remedio"])
resultado_tukey = mc.tukeyhsd()
print(resultado_tukey) 
```

```python
resultado_tukey.plot_simultaneous()
```
