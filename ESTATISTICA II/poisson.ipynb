```python
from scipy.stats import poisson

```

```python
#MEDIA DE ACIDENTES DE CARRO = 2 POR DIA

#qual a prob de ocorrerem 3 acidentes em um dia?
poisson.pmf(3, 2)
```

```python
#qual a prob de ocorrerem 3 ou menos acidentes em um dia?
poisson.cdf(3, 2)
```

```python
#qual a prob de ocorrerem mais de 3 acidentes em um dia?
poisson.sf(3, 2)
```
