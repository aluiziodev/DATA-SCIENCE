```python
import numpy as np
from scipy import stats 

```

```python
jogadores = [40000, 18000, 12000, 30000, 140000, 250000, 300000, 40000, 800000]
print(np.mean(jogadores)) 
```

```python
print(np.median(jogadores))
```

```python
quartis = np.percentile(jogadores, [0, 25, 50, 75, 100])
print(quartis)
```

```python

print(np.std(jogadores, ddof=1))
```
