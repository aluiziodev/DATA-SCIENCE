```python
from scipy.stats import norm
```

```python
#seja uma variável aleatória X que segue uma distribuição normal com média 8 e desvio padrão 2
#qual a probabilidade de X ser menor que 6?

print(norm.cdf(6, 8, 2))
```

```python
#qual a probabilidade de X ser maior que 6?
print(1 - norm.cdf(6, 8, 2))

```

```python
#qual a probabilidade de X ser menor que 6 ou maior que 10?
print(norm.cdf(6, 8, 2) + norm.sf(10, 8, 2))
```

```python
#qual a probabilidade de X ser menor que 10 e maior que 8?
print(norm.cdf(10, 8, 2) - norm.cdf(8, 8, 2))
```
