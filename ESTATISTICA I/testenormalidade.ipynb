```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
db = norm.rvs(size = 1000)
db
```

```python
plt.hist(db, bins=20)
plt.title("dados")
```

```python
fig, ax = plt.subplots()
stats.probplot(db, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(db)
```

```python
#dados nao normais
db2 = skewnorm.rvs(4, size=1000)
print(db2)
```

```python
plt.hist(db2, bins=20)
plt.title("dados nao normais")
```

```python
fig, ax = plt.subplots()
stats.probplot(db2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(db2)
```
