```python
import matplotlib.pyplot as plt
import nltk
nltk.download("stopwords")
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
corpus = PlaintextCorpusReader("Arquivos", ".*", encoding="ISO-8859-1")
```

```python
arquivos = corpus.fileids()
arquivos[0]
```

```python
arquivos[0:10]
```

```python
text = corpus.raw()
```

```python
palavras = corpus.words()
len(palavras)
```

```python
stops = stopwords.words("english")
stops
```

```python
mapa_cores = ListedColormap(["orange", "green", "red", "magenta"])
nuvem = WordCloud(background_color="white", 
                  colormap = mapa_cores, 
                  stopwords = stops,
                  max_words = 100)
nuvem.generate(text)
plt.imshow(nuvem)
```

```python
palavras_semstop = [p for p in palavras if p not in stops]
len(palavras_semstop)
```

```python
palavras_sempont = [p for p in palavras_semstop if p not in string.punctuation]
len(palavras_sempont)
```

```python
freq = nltk.FreqDist(palavras_sempont)
freq
```

```python
mais_comuns = freq.most_common(100)
mais_comuns
```
