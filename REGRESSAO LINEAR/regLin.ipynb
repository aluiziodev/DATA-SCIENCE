```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot
```

```python
db = pd.read_csv('cars.csv')
db.shape
```

```python
db.head()
```

```python
db = db.drop(["Unnamed: 0"], axis=1)
db.head()
```

```python
x = db.iloc[:,1].values
y = db.iloc[:,0].values
x 
```

```python
correlation = np.corrcoef(x, y)
correlation
```

```python
x = x.reshape(-1,1)
modelo = LinearRegression()
modelo.fit(x, y)

```

```python
modelo.intercept_
#interceptaçao
```

```python
modelo.coef_
#inclinaçao 
```

```python
plt.scatter(x, y, color='blue')
plt.plot(x, modelo.predict(x), color='red')
```

```python
# qual a velocidade se levou 22 pes para parar?
modelo.intercept_ + modelo.coef_ * 22
```

```python
modelo.predict([[22]])
```

```python
visualizer = ResidualsPlot(modelo)
visualizer.fit(x, y)
visualizer.poof()
```
