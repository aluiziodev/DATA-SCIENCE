```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
```

```python
db = pd.read_csv('mt_cars.csv')
db.shape
```

```python
db.head()
```

```python
db = db.drop(['Unnamed: 0'], axis=1)
db.head()
```

```python
x = db.iloc[:, 2].values
y = db.iloc[:, 0].values
correlação = np.corrcoef(x, y)
correlação
```

```python
x = x.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(x, y)
```

```python
modelo.intercept_ # b0
```

```python
modelo.coef_ # angular b1
```

```python
modelo.score(x, y) # R²
```

```python
prev = modelo.predict(x)
prev
```

```python
modelo_ajustado = sm.ols(formula='mpg ~ disp', data=db)
modelo_treinado = modelo_ajustado.fit()
modelo_treinado.summary()
```

```python
plt.scatter(x, y, color='blue')
plt.plot(x, prev, color='red')
```

```python
modelo.predict([[200]])
```

```python
x1 = db.iloc[:, 1:4].values
x1
```
